```python
import pymc as pm
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import arviz as az

import logging # dir(logging) for available functionality
_log = logging.getLogger("pymc")
_log.setLevel(logging.INFO) # 20
```

```python
m=1000
true_theta, true_tau = 0, 1
x_observations = stats.norm.rvs(size=m, loc=true_theta, scale=1/true_tau)
```

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*



```python
conjugate_normal_gamma = pm.Model()
with conjugate_normal_gamma:
    theta0 = 0
    theta_prior_n = 1
    tau_SS_prior = 1
    tau_prior_n = 1

    tau0 = theta_prior_n*true_tau


    theta = pm.Normal("theta", mu=theta0, sigma=tau0**(-0.5))
    tau = pm.Gamma("tau", alpha=tau_prior_n/2,
                            beta=tau_SS_prior/2)
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x_observations)

with conjugate_normal_gamma:
    idata_normal_gamma = pm.sample(draws=m, chains=2, tune=100)
```

```python
az.plot_forest(idata_normal_gamma, var_names=["theta", "tau"],
               combined=True, hdi_prob=0.95, r_hat=True)
```

```python
az.plot_trace(idata_normal_gamma, var_names=["theta", "tau"])
```

```python
az.summary(idata_normal_gamma, round_to=5)#2)
```

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*


```python
conjugate_student_t_truncnorm = pm.Model()

with conjugate_student_t_truncnorm:
  theta0 = 0
  theta_prior_n = 1
  tau0 = theta_prior_n * true_tau

  theta = pm.StudentT('theta', mu=theta0, sigma=tau0**(-0.5), nu=2)

  tau_ss_prior = 1
  tau_prior_n = 1
  s = 5
  tau = pm.TruncatedNormal('tau', lower = -tau0/s, upper = np.inf, mu = tau0, sigma = s)

  # the likelihood
  x_obs = pm.Normal('x_obs', mu=theta, sigma=1/tau, observed=x_observations)
```

```python
m=1000
with conjugate_student_t_truncnorm:
  idata_t_norm = pm.sample(draws=m, chains=2, tune=100)
```

```python
az.plot_forest(idata_t_norm, var_names=['theta', 'tau'], combined=True, hdi_prob=0.95, r_hat=True)
```

```python
az.plot_trace(idata_t_norm, var_names=["theta", "tau"])
```

```python
az.summary(idata_t_norm, round_to=5)#2)
```

Provide diagnostic assessments of the performance of your algorithm.*


```python
conjugate_laplace_exp = pm.Model()
with conjugate_laplace_exp:
    theta0 = 0
    theta_prior_n = 1
    tau0 = theta_prior_n * true_tau

    theta = pm.Laplace("theta", mu=theta0, b=1/tau0)

    tau = pm.Exponential("tau", lam=1)

    x_obs = pm.Normal('x_obs', mu=theta, sigma=1/tau, observed=x_observations)

with conjugate_laplace_exp:
    idata_laplace_exp = pm.sample(draws=m, chains=2, tune=100)
```

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):

    # find interval of grid points where f(x_grid) > y
    # then extend the enterval so endpoints f(a)<y and f(b)<y
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid

    x_ = a + stats.uniform().rvs()*(b-a)
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f"
    elif x_ < x:  # or if "x_ was above f on the left side of the interval"
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):

    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)

    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))

m = 100
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):

    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')

    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')

plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```


1. How the Markov algorithm of slice sampling works.

The key idea of slice sampling is that it is sampling from the target distribution f(x) without specifying proposal distribution. The following are the steps for the slice sampling:

    - Given the current state x_t, we sample y ~ normal(0, f(x_t)) to define a horizontal slice.
    - Then, we determine the sampling interval
    - sample a new x within slice.

`slice_f_at_y` is the main function that finds the initial interval and extends the interval. If the sample is within the interval, it accepts, else it will call `slice_f_at_y_` to recursively sampling until the new sample is accepted.

2. Using Slice Sampling in a Metropolis-Within-Gibbs Algorithm

We can replace the Metropolis-Hastings step with slice sampling to adaptively sample from the target distribution. Instead of proposing a new value and accepting/rejecting as in Metropolis-Hastings, slice sampling first draws a horizontal threshold y~unif() to define the slice. The curve represents the unnormalized full conditional distribution of the variable x given others. The initial value for slice sampling is at $x^{(t)}$ and each step is the t+1 that the sampler has to take.

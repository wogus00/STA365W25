### Question 1 part 1
$$\begin{aligned}
p(\boldsymbol{\beta}|\boldsymbol{y}, \boldsymbol{X},{\Sigma}) &\propto \exp\left\{ -\frac{1}{2}(\boldsymbol{y}-\boldsymbol{X\beta})^{\intercal}\Sigma^{-1}(\boldsymbol{y}-\boldsymbol{X\beta})\right\}\cdot p(\boldsymbol{\beta})\\
\text{Where: } p(\boldsymbol{\beta}) &\propto \exp\left\{ -\frac{1}{2}(\boldsymbol{\beta}-\boldsymbol{\beta}_0)^{\intercal}\Sigma^{-1}_{\beta}(\boldsymbol{\beta}-\boldsymbol{\beta}_0)\right\}\\
\implies p(\boldsymbol{\beta}|\boldsymbol{y}, \boldsymbol{X},{\Sigma}) &\propto \exp\left\{ -\frac{1}{2}(\boldsymbol{y}-\boldsymbol{X\beta})^{\intercal}\Sigma^{-1}(\boldsymbol{y}-\boldsymbol{X\beta})\right\}\cdot\exp\left\{ -\frac{1}{2}(\boldsymbol{\beta}-\boldsymbol{\beta}_0)^{\intercal}\Sigma^{-1}_{\beta}(\boldsymbol{\beta}-\boldsymbol{\beta}_0)\right\}\\
&= \exp\left\{ -\frac{1}{2}\left(\boldsymbol{y}^{\intercal}\Sigma^{-1}\boldsymbol{y} - 2\boldsymbol{\beta}^{\intercal}\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{y} + \boldsymbol{\beta}^{\intercal}\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{X\beta}             \right)\right\}
\cdot\exp\left\{ -\frac{1}{2}\left(
  \boldsymbol{\beta}^{\intercal}\Sigma^{-1}_{\beta}\boldsymbol{\beta} - 2\boldsymbol{\beta}_0^{\intercal}\Sigma^{-1}_\beta\boldsymbol{\beta} + \boldsymbol{\beta}_0^{\intercal}\Sigma^{-1}_\beta\boldsymbol{\beta}_0
\right)\right\}\\
&\propto \exp \left\{-\frac{1}{2}\left(
  - 2\boldsymbol{\beta}^{\intercal}\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{y} + \boldsymbol{\beta}^{\intercal}\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{X\beta}
  \right)
  \right\}\cdot
  \exp \left \{-\frac{1}{2} \left(
     \boldsymbol{\beta}^{\intercal}\Sigma^{-1}_{\beta}\boldsymbol{\beta} - 2\boldsymbol{\beta}_0^{\intercal}\Sigma^{-1}_\beta\boldsymbol{\beta}
    \right)
    \right\}                      \\
&=\exp \left \{
  -\frac{1}{2}\left(
      - 2\boldsymbol{\beta}^{\intercal}\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{y} + \boldsymbol{\beta}^{\intercal}\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{X\beta} +      \boldsymbol{\beta}^{\intercal}\Sigma^{-1}_{\beta}\boldsymbol{\beta} - 2\boldsymbol{\beta}_0^{\intercal}\Sigma^{-1}_\beta\boldsymbol{\beta}
    \right)
  \right\}\\
&=\exp \left \{
  -\frac{1}{2}\left(
      - 2\boldsymbol{\beta}^{\intercal}\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{y} + \boldsymbol{\beta}^{\intercal}\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{X\beta} +      \boldsymbol{\beta}^{\intercal}\Sigma^{-1}_{\beta}\boldsymbol{\beta} - 2\boldsymbol{\beta}^{\intercal}\Sigma^{-1}_\beta\boldsymbol{\beta}_0
    \right)
  \right\}\\
&= \exp \left \{-\frac{1}{2}\left[
  \boldsymbol{\beta}^{\intercal}\left(\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{X} + \Sigma^{-1}_\beta\right)\boldsymbol{\beta} -
  2\boldsymbol{\beta}^{\intercal} \left(\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{y} + \Sigma^{-1}_\beta \boldsymbol{\beta}_0\right)
  \right]
  \right \}\\
\text{Here, Let: }
\boldsymbol{Q} &= \left(
  \boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{X} + \Sigma^{-1}_\beta
\right)\\
r &= \left(
\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{y} + \Sigma^{-1}_\beta\boldsymbol{\beta}_0
  \right)\\
\implies p(\boldsymbol{\beta}|\boldsymbol{y}, \boldsymbol{X},{\Sigma}) &\propto \exp \left\{
  -\frac{1}{2} \left(
    \boldsymbol{\beta}^{\intercal}\boldsymbol{Q}\boldsymbol{\beta} - 2\boldsymbol{\beta}^\intercal\boldsymbol{r}
    \right)
  \right\}\\
&=\exp \left\{  -\frac{1}{2}\left[\left(\boldsymbol{\beta} - \boldsymbol{Q}^{-1}\boldsymbol{r}\right)^{\intercal}\boldsymbol{Q}\left(\boldsymbol{\beta}- \boldsymbol{Q}^{-1}\boldsymbol{r}\right) - \boldsymbol{r}^{\intercal}\boldsymbol{Q}^{-1}
\boldsymbol{r}\right]
  \right\}\\
&=\exp \left\{  -\frac{1}{2}\left[
\left(\boldsymbol{\beta} - \boldsymbol{Q}^{-1}\boldsymbol{r}\right)^{\intercal}\boldsymbol{Q}\left(\boldsymbol{\beta}-\boldsymbol{Q}^{-1}\boldsymbol{r}\right)
      \right]
  \right\}\\
&\sim \mathcal{MVN}(\boldsymbol{Q}^{-1}\boldsymbol{r}, \boldsymbol{Q})\\
\text{Where: } \boldsymbol{Q}^{-1}\boldsymbol{r} &= \left(\boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{X} + \Sigma^{-1}_\beta\right)^{-1} \cdot \boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{y} + \Sigma^{-1}_\beta\boldsymbol{\beta}_0\\
\text{So: } \mathcal{Cov}(\boldsymbol{\beta}) &= \boldsymbol{Q} =   \boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{X} + \Sigma^{-1}_\beta\\
\mathbb{E}[\boldsymbol{\beta}] &= [\mathcal{Cov}(\boldsymbol{\beta})]^{-1} \cdot \boldsymbol{X}^{\intercal}\Sigma^{-1}\boldsymbol{y} + \Sigma^{-1}_\beta\boldsymbol{\beta}_0
\end{aligned}$$

#### Normal Prior $\to$ Ridge
$$\begin{aligned}
p(\beta|\boldsymbol{y}) &\propto p(\boldsymbol{y}|\beta)\cdot p(\beta)\\
&=\exp\left\{
  -\frac{1}{2}\left\lvert\lvert
\boldsymbol{y}-X\beta\right\rvert\rvert^2_2
  \right\}\cdot\exp \left\{
  -\frac{1}{2\sigma^2}\left\lvert\lvert {\beta}\right\rvert\rvert^2_2
  \right\}\\
-\log\left(p(\beta|\boldsymbol{y}) \right) &\propto -\log\left(
  \exp\left\{
  -\frac{1}{2}\left\lvert\lvert
\boldsymbol{y}-X\beta\right\rvert\rvert^2_2
  \right\}\cdot\exp \left\{
  -\frac{1}{2\sigma^2}\left\lvert\lvert {\beta}\right\rvert\rvert^2_2
  \right\}
  \right)\\
&= \frac{1}{2}\left\lvert\lvert
\boldsymbol{y}-X\beta\right\rvert\rvert^2_2 +\frac{1}{2\sigma^2}\left\lvert\lvert {\beta}\right\rvert\rvert^2_2
\end{aligned}$$
Where the penalty coefficient $\lambda_{ridge} = \frac{1}{2\sigma^2}$

#### Laplace Prior $\to$ LASSO
$$\begin{aligned}
p(\beta|\boldsymbol{y}) &\propto p(\boldsymbol{y}|\beta)\cdot p(\beta)\\
&=\exp\left\{
  -\frac{1}{2}\left\lvert\lvert
\boldsymbol{y}-X\beta\right\rvert\rvert^2_2
  \right\}\cdot  \exp \left\{
  -\lambda\left\lvert\lvert {\beta}\right\rvert\rvert_1\right\} \\
-\log\left(p(\beta|\boldsymbol{y}) \right) &\propto
-\log\left(
\exp\left\{
  -\frac{1}{2}\left\lvert\lvert
\boldsymbol{y}-X\beta\right\rvert\rvert^2_2
  \right\}\cdot  \exp \left\{
  -\lambda\left\lvert\lvert {\beta}\right\rvert\rvert_1\right\}
    \right)\\
&=  \frac{1}{2}\left\lvert\lvert
\boldsymbol{y}-X\beta\right\rvert\rvert^2_2 + \lambda\left\lvert\lvert {\beta}\right\rvert\rvert_1
\end{aligned}$$
Where penalty coefficient $\lambda_{lasso} = \lambda$ from the Laplace distribution.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import pymc as pm
import arviz as az
```

```python
tips = sns.load_dataset("tips")
X_data = tips["total_bill"].values
y_data = tips["tip"].values

# make design matrix for X
X = np.column_stack([np.ones_like(X_data), X_data])
```

```python
with pm.Model() as model:
  p = X.shape[1]
  betas = pm.Normal("betas", mu=0, sigma=1, shape=p)
  sigma = pm.HalfNormal("sigma", sigma=1)
  mu = pm.math.dot(X,betas)

  observations = pm.Normal('y_obs', mu=mu, sigma=sigma, observed = y_data)

  trace = pm.sample(
      draws=2000,
      chains=4
      )
  ppc = pm.sample_posterior_predictive(trace, var_names=["y_obs"])
```

```python
summary_table = az.summary(trace, var_names=["betas", "sigma"], round_to=3)
print(summary_table)
az.plot_trace(trace, var_names=["betas", "sigma"]);
```

#### Bayesian Posterior Analysis

We see that $\mathbb{E}[\beta_0] = 0.893, \mathcal{sd}(\beta_0) = 0.156$, with 94% credible inverval of $[0.610, 1.196]$.

We see that $\mathbb{E}[\beta_1] = 0.106, \mathcal{sd}(\beta_1) = 0.007$, with 94% credible inverval of $[0.092, 0.119]$.

We see that $\mathbb{E}[\sigma] = 1.025, \mathcal{sd}(\sigma) = 0.046$, with 94% credible inverval of $[0.940, 1.112]$.



#### MCMC Diagnostics

From the summary table, we know see that `r_hat` for all variables are close to 1.0. This indicates that between-chain variance and within-chain variance are similar, which suggests good convergence.

Also, we see that `ess_bulk` and `ess_tail`are above 2000 for all variables, which indicates a large effective sample size. This tells us that there is low autocorrelation in the chains and that the posterior estiamtes are based on many independent draws.

```python
# 1) Load tips dataset
tips = sns.load_dataset("tips")

# 2) Convert to numpy arrays
X_data = tips["total_bill"].values
y_data = tips["tip"].values
n = len(y_data)

rng = np.random.default_rng(42)
outlier_indices = rng.choice(n, size=5, replace=False)

for i in outlier_indices:
    shift = rng.choice([2, 4]) * rng.uniform(5, 10)
    y_data[i] = y_data[i] * shift
X = np.column_stack([np.ones(n), X_data])
print("Outlier indices:", outlier_indices)
true_outlier_indices = sorted(outlier_indices)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_data, y_data, alpha=0.6, label="Data")
plt.scatter(X_data[outlier_indices], y_data[outlier_indices], color="red", label="Outliers")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Tips Data with Artificial Outliers")
plt.legend()
plt.show()
```

```python
with pm.Model() as robust_model:

    # Priors on regression coefficients
    beta0 = pm.Normal("beta0", mu=0, sigma=10)
    beta1 = pm.Normal("beta1", mu=0, sigma=10)

    nu = pm.Exponential("nu", 1.0)

    alpha = nu / 2
    beta = nu / 2

    tau = pm.Gamma("tau", alpha=alpha, beta=beta, shape=n)

    mu = beta0 + beta1 * X_data

    y_obs = pm.Normal("y_obs", mu=mu, tau=tau, observed=y_data)

    # MCMC sampling
    trace = pm.sample(
        draws=3000,
        chains=4,
    )
```

```python
tau_samples = trace.posterior["tau"]

tau_samples_stacked = tau_samples.stack(samples=("chain", "draw"))

sigma2_samples_stacked = 1.0 / tau_samples_stacked

tau_mean = tau_samples_stacked.mean(dim="samples").values
sigma2_mean = sigma2_samples_stacked.mean(dim="samples").values

sorted_ix = np.argsort(sigma2_mean)[::-1]

n_outliers = 5
outlier_indices = sorted_ix[:n_outliers]


print("True Outlier Indicies:", true_outlier_indices)
print("Outlier candidates:", outlier_indices)
print("Posterior mean sigma2:", sigma2_mean[outlier_indices])
```

```python
X_mat = np.column_stack([np.ones(n), X_data])

tau_samples_stacked = tau_samples.stack(sample=("chain", "draw")).transpose("sample", "tau_dim_0")
tau_samples_arr = tau_samples_stacked.values

n_samples = tau_samples_arr.shape[0]
H_diags = np.zeros((n_samples, n))

for d in range(n_samples):
    W_d = np.diag(tau_samples_arr[d, :])
    sqrtW_d = np.sqrt(W_d)
    X_weighted = sqrtW_d @ X_mat
    XtWX_inv = np.linalg.inv(X_weighted.T @ X_weighted)
    H_d = X_weighted @ XtWX_inv @ X_weighted.T
    H_diags[d, :] = np.diag(H_d)

H_mean = H_diags.mean(axis=0)  # shape (n,)
H_sd   = H_diags.std(axis=0)

high_influence_indices = np.where(H_mean > 2*2/n)[0]  # e.g. threshold 0.2 or 2*(p/n), etc.
print("High-influence points:", high_influence_indices)
print("Mean hat values of those points:", H_mean[high_influence_indices])
```

```python
print("influence of outlier indicies:")
for idx in true_outlier_indices:
  print(f"    index {idx}: {H_diags.mean(axis=0)[int(idx)]}")

high_influence_indices = np.where(H_diags.mean(axis=0) > 2*2/n)[0]
print("high influece indicies:")
for idx in high_influence_indices:
  print(f"    index {idx}: {H_diags.mean(axis=0)[int(idx)]}")
```

```python
# Posterior samples of tau in shape (n_samples, n)
tau_samples_arr = tau_samples.stack(sample=("chain", "draw")).transpose("sample", "tau_dim_0").values

# We'll compare the posterior distribution of sigma^2 = 1/tau for an outlier vs. a non-outlier.
outlier_idx = true_outlier_indices[0]
non_outlier_idx = 0

sigma2_outlier_samples = 1.0 / tau_samples_arr[:, outlier_idx]
sigma2_non_outlier_samples = 1.0 / tau_samples_arr[:, non_outlier_idx]

print(f"Comparing point {outlier_idx} (outlier) vs. {non_outlier_idx} (non-outlier)")

print("Outlier index", outlier_idx)
print("  sigma^2 mean:", sigma2_outlier_samples.mean())
print("  sigma^2 std :", sigma2_outlier_samples.std())
print()
print("Non-outlier index", non_outlier_idx)
print("  sigma^2 mean:", sigma2_non_outlier_samples.mean())
print("  sigma^2 std :", sigma2_non_outlier_samples.std())
```

The results clearly show that point 21, which was as an outlier, has a much larger posterior mean for $\sigma^2$ (about 5345) compared to the non‐outlier point 0 (about 4.73). This large variance indicates that the model effectively “down‐weights” point 21 by allowing it far more uncertainty, consistent with it being an outlier. Meanwhile, point 0 is well‐explained by a relatively small variance, suggesting it follows the typical trend in the data.
